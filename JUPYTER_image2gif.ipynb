```python
%reset -f

import imageio
import os
```

```python
#CurDir = r'C:\temp\Footage\gif'
CurDir = './images'
CurFiles = []
print('Current file path: ', CurDir)
for idx, file in enumerate(sorted(os.listdir(os.path.join(CurDir)))):
    if (file.endswith("g")): # Maybe use more filters--> file.endswith(".jpg") & ('xyz' in file)
        CurFiles.append(file)
        
#sort files according to number in filename if needed
CurFiles.sort(key=lambda f: int(''.join(filter(str.isdigit, f))))

for i, fl in enumerate(CurFiles):
    print(i, ': ', fl)
```

```python
with imageio.get_writer(os.path.join(CurDir, 'output2.gif'), mode='I', fps=3) as writer:
    for filename in CurFiles:
        image = imageio.imread(os.path.join(CurDir, filename))
        writer.append_data(image)
```
